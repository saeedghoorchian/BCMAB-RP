```python
import pickle
import matplotlib.pyplot as plt
```

```python
with open('results/results_t_100_n_1_d_68.pickle', "rb") as f:
    results = pickle.load(f)
```

```python
results[2]
```

```python
def plot_ctrs(ctr_dict):
    fig, ax = plt.subplots(1,1, figsize=(10, 5));
    
    N = 40
    
    for label, ctr in ctr_dict.items():
        ax.plot(ctr[N:], label=label)
    ax.set_xlabel('Trial')
    ax.set_ylabel('CTR')
    ax.set_title('Evaluation without costs')

    plt.legend();
    plt.show();
```

```python
ctr_dict = {key: value[-1] for key,value in results[2].items()}
```

```python
plot_ctrs(ctr_dict)
```
